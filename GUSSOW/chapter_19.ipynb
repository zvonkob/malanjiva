```python
from math import *
import matplotlib.pyplot as plt
```

```python
# ex.19.2

L = 20
R = 20
U = 120

def i_func(I, R, L, t):
    return I * (1 - exp(- (R / L) * t))

I = U / R

tt = []
ii = []
for t_ in range(50):
    t = t_ / 10
    i = i_func(I, R, L, t)
    tt.append(t)
    ii.append(i)
    i = round(i, 2)
    if t_ % 10 == 0:
        print(t, 's,', i, 'A')

plt.title('Waveform i(t)')
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.plot(tt, ii)
```

```python
# ex.19.2

def uR_func(U, R, L, t):
    return U * (1 - exp(- (R / L) * t))

tt = []
uu = []
for t_ in range(50):
    t = t_ / 10
    u = uR_func(U, R, L, t)
    tt.append(t)
    uu.append(u)
    u = round(u, 2)
    if t_ % 10 == 0:
        print(t, 's,', u, 'V')

plt.title('Waveform uR(t)')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.plot(tt, uu, color='red')
```

```python
# ex.19.2

def uL_func(U, R, L, t):
    return U * exp(- (R / L) * t)

tt = []
uu = []
for t_ in range(50):
    t = t_ / 10
    u = uL_func(U, R, L, t)
    tt.append(t)
    uu.append(u)
    u = round(u, 2)
    if t_ % 10 == 0:
        print(t, 's,', u, 'V')

plt.title('Waveform uR(t)')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.plot(tt, uu, color='violet')
```
